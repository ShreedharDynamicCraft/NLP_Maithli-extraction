```python
# Wikipedia pages HTML format mein hote hain. Hum sirf <div class="mw-parser-output"> for content
```

```python
from bs4 import BeautifulSoup
import os

input_dir = "downloaded_site/mai.wikipedia.org/wiki"
output_dir = "text_files"
os.makedirs(output_dir, exist_ok=True)

for filename in os.listdir(input_dir):
    if filename.endswith(".html"):
        with open(os.path.join(input_dir, filename), "r", encoding="utf-8") as f:
            soup = BeautifulSoup(f, "html.parser")
            content = soup.find("div", class_="mw-parser-output")
            if content:
                text = content.get_text(separator="\n")
                with open(os.path.join(output_dir, filename.replace(".html", ".txt")), "w", encoding="utf-8") as out:
                    out.write(text)

print(f"Text files created in: {output_dir}")
print("Sample files:", os.listdir(output_dir)[:5])
```

```python
import glob

text_files = glob.glob("text_files/*.txt")
with open("corpus.txt", "w", encoding="utf-8") as outfile:
    for fname in text_files:
        with open(fname, "r", encoding="utf-8") as infile:
            outfile.write(infile.read() + "\n")

print("Corpus created: corpus.txt")
print(f"Size of corpus: {os.path.getsize('corpus.txt') / 1024:.2f} KB")
```
