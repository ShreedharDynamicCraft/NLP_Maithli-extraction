```python
from bs4 import BeautifulSoup
import os
import re

input_dir = "downloaded_site/mai.wikipedia.org/wiki"
output_dir = "text_files"
os.makedirs(output_dir, exist_ok=True)

for filename in os.listdir(input_dir):
    if filename.endswith(".html"):
        filepath = os.path.join(input_dir, filename)
        try:
            with open(filepath, "r", encoding="utf-8") as f:
                soup = BeautifulSoup(f, "html.parser")
                content = soup.find("div", class_="mw-parser-output")
                if content:
                    # Remove unwanted sections
                    for unwanted in content(["table", "sup", "h1", "h2", "h3", "div"]):
                        unwanted.decompose()
                    text = content.get_text(separator="\n").strip()
                    # Clean up references and edit links
                    text = re.sub(r"\[\d+\]", "", text)  # Remove [1], [2], etc.
                    text = re.sub(r"\[सम्पादन करी\]", "", text)  # Remove [edit]
                    text = "\n".join(line.strip() for line in text.splitlines() if line.strip())  # Remove empty lines
                    if text and len(text) > 100:  # Meaningful text only
                        with open(os.path.join(output_dir, filename.replace(".html", ".txt")), "w", encoding="utf-8") as out:
                            out.write(text)
                            print(f"Processed: {filename}")
        except Exception as e:
            print(f"Error in {filename}: {e}")

txt_files = os.listdir(output_dir)
print(f"Total text files created: {len(txt_files)}")
print("Sample text files:", txt_files[:5])
```

```python
import re

# Maithili Morphological Analyzer (Corrected Final Version)
def analyze_word(word):
    # Dictionary of irregular words
    irregular_words = {
        "हम": "Pronoun (1st Person)",
        "तोरा": "Pronoun (2nd Person)",
        "गेल": "Verb (Past)",
        "आबि": "Verb (Come)",
        "छी": "Verb (Present, 3rd Person)",
        "अछ": "Verb (Present, 3rd Person)"
    }

    # Check irregular words first
    if word in irregular_words:
        return f"{word}: {irregular_words[word]}"

    # Morphological Rules (Specific to Generic order)
    if word.endswith("कर"):  # Action verb - check before postpositions
        return f"{word}: Verb (Action)"
    elif word.endswith("छी") or word.endswith("अछ"):  # Present tense verb
        return f"{word}: Verb (Present)"
    elif word.endswith("लौं") or word.endswith("लहुं"):  # Past tense verb
        return f"{word}: Verb (Past)"
    elif word.endswith("आ") or word.endswith("इ"):  # Noun endings
        return f"{word}: Noun"
    elif word.endswith("बा") or word.endswith("बे"):  # Adjective endings
        return f"{word}: Adjective"
    elif word.endswith("एर") or word.endswith("आर"):  # Possessive case
        return f"{word}: Possessive Pronoun"
    elif word.endswith("सभ"):  # Plural marker
        return f"{word}: Noun (Plural)"
    elif re.match(r"^अन", word):  # Negation prefix
        return f"{word}: Negative Prefix Word"
    elif word in ["क", "म", "स"]:  # Common prepositions/postpositions - check last
        return f"{word}: Postposition"
    else:
        return f"{word}: Unknown"

# Read words from cleanCorpus.txt
with open("cleanCorpus.txt", "r", encoding="utf-8") as f:
    text = f.read().split()

# Analyze and save results
with open("morph_analysis.txt", "w", encoding="utf-8") as f:
    for word in text:
        f.write(analyze_word(word) + "\n")

# Print first 10 analyzed words
print("First 10 Analyzed Words from cleanCorpus.txt:")
with open("morph_analysis.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
    for i, line in enumerate(lines[:10], 1):
        print(f"{i}. {line.strip()}")

print("Morphological analysis written to morph_analysis.txt")
```

Step 9: Removeing  Stop Words

"क", "म", "स", "अछ", "छी", "के", "से", "आ", "जे", "ह", "ओ", "ए", "एक", "अपने", "हम", "त", "नहि", "छ", "हु", "तू".
Yeh Maithili ke common postpositions ("क", "म", "से"), pronouns ("हम", "तू"), verbs ("अछ", "छी"), aur negation ("नहि") ko cover karta .

```python
import sys

# Maithili Stop Words List (Further Expanded)
stop_words = {
    "क", "म", "स", "अछ", "छी", "के", "से", "आ", "जे", "ह", 
    "ओ", "ए", "एक", "अपने", "हम", "त", "नहि", "छ", "हु", "तू",
    "ल", "बाद", "सभ", "कि", "पर", "ही", "बि", "न", "ता", "रे",
    "ई", "इस", "एहि", "का", "की", "को", "ने", "द्वारा", "लागि", "बेर"
}

def remove_stopwords(input_file, output_file):
    """Removes stop words from the given text file and saves the cleaned text."""
    try:
        with open(input_file, "r", encoding="utf-8") as f:
            sentences = f.readlines()
    except FileNotFoundError:
        print(f"❌ Error: Input file '{input_file}' not found!")
        sys.exit(1)
    except Exception as e:
        print(f"❌ Error reading file: {e}")
        sys.exit(1)

    cleaned_sentences = []
    for sentence in sentences:
        words = sentence.strip().split()
        cleaned_words = [word for word in words if word not in stop_words]
        if cleaned_words:
            cleaned_sentences.append(" ".join(cleaned_words))

    try:
        with open(output_file, "w", encoding="utf-8") as f:
            f.write("\n".join(cleaned_sentences) + "\n")
    except Exception as e:
        print(f"❌ Error writing to file: {e}")
        sys.exit(1)

    return cleaned_sentences

# File Processing (Using cleanCorpus.txt instead of sentences.txt)
try:
    output_sentences = remove_stopwords("cleanCorpus.txt", "no_stopwords.txt")
except SystemExit:
    sys.exit(1)

# Print first 10 cleaned lines
print("First 10 Lines of no_stopwords.txt:")
for i, line in enumerate(output_sentences[:10], 1):
    print(f"{i}. {line}")

print("\n✅ Stop words removed and saved to no_stopwords.txt")
print("📜 Stop Words List Used:", stop_words)
```
